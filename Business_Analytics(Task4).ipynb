##**K-Means Clustering**

```python
import pandas as pd
```

```python
ratings = [['john',5,5,2,1],['mary',4,5,3,2],['bob',4,4,4,3],['lisa',2,2,4,5],['lee',1,2,3,4],['harry',2,1,5,5]]
titles = ['user','Jaws','Star Wars','Exorcist','Omen']
movies = pd.DataFrame(ratings,columns=titles)
movies
```

```python
#applying k-means clustering (with k=2) on the movie ratings data
#The "user" column is first removed before applying the clustering algorithm

from sklearn import cluster 

new_data = movies.drop('user',axis=1)
k_means = cluster.KMeans(n_clusters=2, max_iter=50, random_state=1)
k_means.fit(new_data) 
labels = k_means.labels_
pd.DataFrame(labels, index=movies.user, columns=['Cluster_ID'])
```
